```python
import requests
from bs4 import BeautifulSoup
import csv
import urllib.parse
import re
from datetime import datetime

# Define the URLs for the best movies list
urls = [
     "https://editorial.rottentomatoes.com/guide/essential-movies-to-watch-now/",
    "https://editorial.rottentomatoes.com/guide/essential-movies-to-watch-now/2/",
    "https://editorial.rottentomatoes.com/guide/freshest-movies-past-20-years/",
    "https://editorial.rottentomatoes.com/guide/140-essential-action-movies-to-watch-now/",
    "https://editorial.rottentomatoes.com/guide/best-movies-of-all-time/",
    "https://editorial.rottentomatoes.com/guide/best-movies-of-all-time/2/",
    "https://editorial.rottentomatoes.com/guide/warner-bros-100-essential-movies/",
    "https://editorial.rottentomatoes.com/guide/max-100-best-movies/",
    "https://editorial.rottentomatoes.com/guide/best-2014-movies/",
    "https://editorial.rottentomatoes.com/guide/best-horror-movies-of-all-time/",
    "https://editorial.rottentomatoes.com/guide/essential-slasher-movies/",
    "https://editorial.rottentomatoes.com/guide/essential-comedy-movies/",
    "https://editorial.rottentomatoes.com/guide/essential-comedy-movies/2/",
    "https://editorial.rottentomatoes.com/guide/best-computer-animated-movies-of-all-time/",
    "https://editorial.rottentomatoes.com/guide/best-war-movies-of-all-time/",
    "https://editorial.rottentomatoes.com/guide/best-war-movies-of-all-time/2/",
    "https://editorial.rottentomatoes.com/guide/best-sports-movie-of-all-time/",
    "https://editorial.rottentomatoes.com/guide/best-comedy-shows-of-all-time/",
    "https://editorial.rottentomatoes.com/guide/best-comedy-shows-of-all-time/2/",
    "https://editorial.rottentomatoes.com/guide/best-comedy-shows-of-all-time/3/",
    "https://editorial.rottentomatoes.com/guide/best-comedy-shows-of-all-time/4/",
    "https://editorial.rottentomatoes.com/guide/best-christmas-movies/",
    "https://editorial.rottentomatoes.com/guide/best-movies-on-amazon-prime-right-now/",
    "https://editorial.rottentomatoes.com/guide/best-new-animated-movies/",
    "https://editorial.rottentomatoes.com/guide/best-thanksgiving-movies/",
    "https://editorial.rottentomatoes.com/guide/best-new-movies/",
    "https://editorial.rottentomatoes.com/guide/best-new-movies/2/",
    "https://editorial.rottentomatoes.com/guide/best-musical-movies-of-all-time/",
    "https://editorial.rottentomatoes.com/guide/best-movies-tubi/",
    "https://editorial.rottentomatoes.com/guide/best-free-youtube-movies/",
    "https://editorial.rottentomatoes.com/guide/best-asian-american-movies/",
    "https://editorial.rottentomatoes.com/guide/oscars-best-and-worst-best-pictures/",
    "https://editorial.rottentomatoes.com/guide/golden-globes-best-film-winners-by-tomatometer/",
    "https://editorial.rottentomatoes.com/guide/best-heist-movies-of-all-time/",
    "https://editorial.rottentomatoes.com/guide/best-fantasy-movies-of-all-time/",
    "https://editorial.rottentomatoes.com/guide/essential-sci-fi-movies-of-all-time/",
    "https://editorial.rottentomatoes.com/guide/essential-animated-movies/",
    "https://editorial.rottentomatoes.com/guide/essential-animated-movies/2/",
    "https://editorial.rottentomatoes.com/guide/best-western-movies/",
    "https://editorial.rottentomatoes.com/guide/best-high-school-movies/",
    "https://editorial.rottentomatoes.com/guide/best-horror-movies-2020s/",
    "https://editorial.rottentomatoes.com/guide/best-horror-movies-2020s/2/",
    "https://editorial.rottentomatoes.com/guide/best-horror-movie-sequels/",
    "https://editorial.rottentomatoes.com/guide/best-2010s-horror-movies/",
    "https://editorial.rottentomatoes.com/guide/best-2000s-horror-movies/",
    "https://editorial.rottentomatoes.com/guide/best-horror-movies-directed-by-women/",
    "https://editorial.rottentomatoes.com/guide/best-anime-movies/",
    "https://editorial.rottentomatoes.com/guide/best-1994-movies/",
    "https://editorial.rottentomatoes.com/guide/essential-zombie-movies/",
    "https://editorial.rottentomatoes.com/guide/best-80s-horror-movies/",
    "https://editorial.rottentomatoes.com/guide/all-dreamworks-animation-movies-ranked/",
    "https://editorial.rottentomatoes.com/guide/best-1970s-horror-movies/",
    "https://editorial.rottentomatoes.com/guide/best-2009-movies/",
    "https://editorial.rottentomatoes.com/guide/best-found-footage-movies/",
    "https://editorial.rottentomatoes.com/guide/all-tom-cruise-movies-ranked/",
    "https://editorial.rottentomatoes.com/guide/movie-duos/",
    "https://editorial.rottentomatoes.com/guide/best-1999-movies/",
    "https://editorial.rottentomatoes.com/guide/best-1950s-sci-fi-movies/",
    "https://editorial.rottentomatoes.com/guide/all-a24-movies-ranked/",
    "https://editorial.rottentomatoes.com/guide/essential-2000s-movies/",
    "https://editorial.rottentomatoes.com/guide/best-1969-movies/",
    "https://editorial.rottentomatoes.com/guide/best-baseball-movies/",
    "https://editorial.rottentomatoes.com/guide/mgm-100-movies/",
    "https://editorial.rottentomatoes.com/guide/all-nicolas-cage-movies-ranked/",
    "https://editorial.rottentomatoes.com/guide/best-dog-movies-ranked-by-tomatometer/",
    "https://editorial.rottentomatoes.com/guide/best-movies-3-hours-or-longer-ranked/",
    "https://editorial.rottentomatoes.com/guide/best-romantic-comedies-of-all-time/",
    "https://editorial.rottentomatoes.com/guide/best-black-movies-21st-century/",
    "https://editorial.rottentomatoes.com/guide/best-1984-movies/",
    "https://editorial.rottentomatoes.com/guide/sad-movies/",
    "https://editorial.rottentomatoes.com/guide/best-movies-of-2023/",
    "https://editorial.rottentomatoes.com/guide/best-horror-movies-2023/",
    "https://editorial.rottentomatoes.com/guide/worst-movies-of-all-time/",
    "https://editorial.rottentomatoes.com/guide/140-favorite-80s-movies/",
    "https://editorial.rottentomatoes.com/guide/best-french-horror-movies/",
    "https://editorial.rottentomatoes.com/guide/best-korean-horror-movies/",
    "https://editorial.rottentomatoes.com/guide/summer-movie-scorecard-2023/",
    "https://editorial.rottentomatoes.com/guide/best-boxing-movies-of-all-time/",
    "https://editorial.rottentomatoes.com/guide/best-movies-written-by-women/",
    "https://editorial.rottentomatoes.com/guide/best-movies-2022/",
    "https://editorial.rottentomatoes.com/guide/2022-horror-movies-ranked/",
    "https://editorial.rottentomatoes.com/guide/rotten-movies-you-loved-2022/",
    "https://editorial.rottentomatoes.com/guide/140-favorite-90s-movies/",
    "https://editorial.rottentomatoes.com/guide/100-best-documentaries/",
    "https://editorial.rottentomatoes.com/guide/toronto-film-festival-2022-movie-scorecard/",
    "https://editorial.rottentomatoes.com/guide/ice-cube/",
    "https://editorial.rottentomatoes.com/guide/best-assassin-movies/",
    "https://editorial.rottentomatoes.com/guide/favorite-1970s-movies/",
    "https://editorial.rottentomatoes.com/guide/favorite-1970s-movies/2/",
    "https://editorial.rottentomatoes.com/guide/favorite-1970s-movies/3/",
    "https://editorial.rottentomatoes.com/guide/cannes-2022-scorecard/",
    "https://editorial.rottentomatoes.com/guide/best-actress-oscar-movies/",
    "https://editorial.rottentomatoes.com/guide/2021-best-movies/",
    "https://editorial.rottentomatoes.com/guide/best-cinematography-winners-ranked/",
    "https://editorial.rottentomatoes.com/guide/sxsw2022-movie-scorecard/",
    "https://editorial.rottentomatoes.com/guide/best-action-movies-of-2021-ranked/",
    "https://editorial.rottentomatoes.com/guide/best-horror-movies-of-2021-ranked/",
    "https://editorial.rottentomatoes.com/guide/best-anime-movies-2021/",
    "https://editorial.rottentomatoes.com/guide/sxsw-2021-movie-scorecard/",
    "https://editorial.rottentomatoes.com/guide/best-action-movies-of-2020/",
    "https://editorial.rottentomatoes.com/guide/best-comedy-movies-of-2020/",
    "https://editorial.rottentomatoes.com/guide/best-horror-movies-of-2020/",
    "https://editorial.rottentomatoes.com/guide/the-best-movies-of-2020/",
    "https://editorial.rottentomatoes.com/guide/best-action-movies-of-2019/",
    "https://editorial.rottentomatoes.com/guide/best-movies-of-2019/",
    "https://editorial.rottentomatoes.com/guide/the-200-best-movies-of-the-2010s/",
    "https://editorial.rottentomatoes.com/guide/best-comedies-2019/",
    "https://editorial.rottentomatoes.com/guide/best-tv-and-streaming-movies-of-2019/",
    "https://editorial.rottentomatoes.com/guide/toronto-international-film-festival-2019-scorecard/",
    "https://editorial.rottentomatoes.com/guide/cannes-2019-movie-scorecard/",
    "https://editorial.rottentomatoes.com/guide/all-fox-searchlight-movies-ranked/",
    "https://editorial.rottentomatoes.com/guide/sxsw-2019-movie-scorecard/",
    "https://editorial.rottentomatoes.com/guide/sundance-2019-scorecard/",
    "https://editorial.rottentomatoes.com/guide/best-horror-movies-2018-tomatometer/",
    "https://editorial.rottentomatoes.com/guide/best-movies-of-2018/",
    "https://editorial.rottentomatoes.com/guide/best-comedy-movies-2018-tomatometer/",
    "https://editorial.rottentomatoes.com/guide/toronto-2018-scorecard/",
    "https://editorial.rottentomatoes.com/guide/sundance-2018-movies-by-tomatometer/",
    "https://editorial.rottentomatoes.com/guide/best-fresh-romantic-movies/",
    "https://editorial.rottentomatoes.com/guide/holiday-movie-scorecard-2017/",
    "https://editorial.rottentomatoes.com/guide/scary-movies-of-2017-by-tomatometer/",
    "https://editorial.rottentomatoes.com/guide/toronto-international-film-festival-2017-scorecard/",
    "https://editorial.rottentomatoes.com/guide/toronto-international-film-festival-2017-scorecard/2/",
    "https://editorial.rottentomatoes.com/guide/toronto-international-film-festival-2017-scorecard/3/",
    "https://editorial.rottentomatoes.com/guide/toronto-international-film-festival-2017-scorecard/4/"
]

# OMDb API keys
OMDB_API_KEYS = ['6a7c97d', 'bf5c1295', 'ae2178ec', '17daa304', '8cdbac1e', '352077f6', '245d1d8f', '9da6c667', 'c92e05fa', '70d5de12', '2f6f7d8f', '9694a667', '87e9cc1e', '3e1217f6', 'a51318ec', '1ce8c304', 'd95a97d', 'b46e7295', 'e0e6b108', '591d6ae0', '8107ec1e']  # Replace with your actual API keys
CURRENT_API_INDEX = 0  # Index of the current API key

# Function to clean movie title
def clean_movie_title(raw_title):
    return re.sub(r'\(\d{4}\)\d*%?', '', raw_title).strip()

# Function to fetch genre from OMDb API using rotating keys
def get_movie_genre(movie_name):
    global CURRENT_API_INDEX
    cleaned_title = clean_movie_title(movie_name)
    
    for _ in range(len(OMDB_API_KEYS)):  # Try each API key once
        api_key = OMDB_API_KEYS[CURRENT_API_INDEX]
        query_url = f"http://www.omdbapi.com/?apikey={api_key}&t={urllib.parse.quote(cleaned_title)}"
        response = requests.get(query_url)

        if response.status_code == 200:
            data = response.json()
            if data.get('Response') == 'True':
                print(f"Success: API key {CURRENT_API_INDEX + 1} provided data for '{cleaned_title}'")
                return data.get('Genre', 'N/A')
            print(f"API key {CURRENT_API_INDEX + 1}: No exact match for '{cleaned_title}'")
        else:
            print(f"Error: API key {CURRENT_API_INDEX + 1} request failed for '{cleaned_title}', Status code {response.status_code}")
        
        # Move to the next API key
        CURRENT_API_INDEX = (CURRENT_API_INDEX + 1) % len(OMDB_API_KEYS)
    
    print(f"All API keys failed for '{cleaned_title}'")
    return 'N/A'

# Function to fetch movie details from OMDb API using rotating keys
def get_movie_details(movie_name):
    global CURRENT_API_INDEX
    cleaned_title = clean_movie_title(movie_name)
    
    for _ in range(len(OMDB_API_KEYS)):  # Try each API key once
        api_key = OMDB_API_KEYS[CURRENT_API_INDEX]
        query_url = f"http://www.omdbapi.com/?apikey={api_key}&t={urllib.parse.quote(cleaned_title)}"
        response = requests.get(query_url)

        if response.status_code == 200:
            data = response.json()
            if data.get('Response') == 'True':
                print(f"Success: API key {CURRENT_API_INDEX + 1} provided data for '{cleaned_title}'")
                
                # Extract Metacritic and IMDb from the Ratings array
                metacritic_rating = 'N/A'
                imdb_rating = 'N/A'
                for rating in data.get('Ratings', []):
                    if rating['Source'] == 'Metacritic':
                        metacritic_rating = rating['Value']
                    if rating['Source'] == 'Internet Movie Database':
                        imdb_rating = rating['Value']
                
                return {
                    'Year': data.get('Year', 'N/A'),
                    'Runtime': data.get('Runtime', 'N/A'),
                    'Rated': data.get('Rated', 'N/A'),
                    'Director': data.get('Director', 'N/A'),
                    'Writer': data.get('Writer', 'N/A'),
                    'Actors': data.get('Actors', 'N/A'),
                    'Plot': data.get('Plot', 'N/A'),
                    'Language': data.get('Language', 'N/A'),
                    'Country': data.get('Country', 'N/A'),
                    'Metacritic': metacritic_rating,
                    'Metascore': data.get('Metascore', 'N/A'),
                    'imdbRating': imdb_rating,
                    'Genre': get_movie_genre(movie_name),  # Added genre here
                }
            print(f"API key {CURRENT_API_INDEX + 1}: No exact match for '{cleaned_title}'")
        else:
            print(f"Error: API key {CURRENT_API_INDEX + 1} request failed for '{cleaned_title}', Status code {response.status_code}")
        
        # Move to the next API key
        CURRENT_API_INDEX = (CURRENT_API_INDEX + 1) % len(OMDB_API_KEYS)
    
    print(f"All API keys failed for '{cleaned_title}'")
    return {}

# Function to scrape movies from Rotten Tomatoes
def scrape_movies():
    movie_data = []
    seen_movies = set()  # Set to track unique movie names

    for url in urls:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            movies = soup.find_all('div', class_='article_movie_title')

            for movie in movies:
                movie_name = movie.get_text(strip=True)

                # Skip if the movie is already in the dataset
                if movie_name in seen_movies:
                    continue

                t_meter_score = movie.find_next('span', class_='tMeterScore').get_text(strip=True)
                movie_details = get_movie_details(movie_name)  # Get additional movie details

                # Append movie data along with the new columns in the requested order
                movie_data.append([
                    movie_name, 
                    movie_details.get('Year', 'N/A'),
                    movie_details.get('Genre', 'N/A'),
                    movie_details.get('Runtime', 'N/A'),
                    movie_details.get('Rated', 'N/A'),
                    movie_details.get('Director', 'N/A'),
                    movie_details.get('Writer', 'N/A'),
                    movie_details.get('Actors', 'N/A'),
                    movie_details.get('Plot', 'N/A'),
                    movie_details.get('Language', 'N/A'),
                    movie_details.get('Country', 'N/A'),
                    t_meter_score, 
                    movie_details.get('Metacritic', 'N/A'),
                    movie_details.get('Metascore', 'N/A'),
                    movie_details.get('imdbRating', 'N/A'),
                ])
                seen_movies.add(movie_name)  # Add movie name to the set
        else:
            print(f"Failed to retrieve: {url}")
    return movie_data

# Function to save movie data to CSV
def save_to_csv(movie_data):
    timestamp = datetime.now().strftime("%d%m%Y_%H%M%S")
    filename = f'best_movies_{timestamp}.csv'
    
    try:
        with open(filename, mode='w', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerow([
                'Movie Name', 'Year', 'Genre', 'Runtime', 'Rated', 'Director', 'Writer',
                'Actors', 'Plot', 'Language', 'Country', 'Tomatometer Score', 'Metacritic', 
                'Metascore', 'imdbRating'  # Removed 'Internet Movie Database' column
            ])
            writer.writerows(movie_data)
        print(f"Data saved to {filename}")
    except PermissionError:
        print(f"PermissionError: Unable to write CSV file '{filename}'.")

# Main execution
if __name__ == "__main__":
    movie_data = scrape_movies()
    save_to_csv(movie_data)
```

```python
import pandas as pd
import numpy as np

# Function to load and clean CSV data
def clean_csv_data(input_file, output_file):
    # Read the CSV file
    df = pd.read_csv(input_file)

    # Replace 'N/A' and other unwanted values with NaN
    df.replace('N/A', np.nan, inplace=True)

    # Strip extra spaces from string columns
    string_columns = ['Movie Name', 'Genre', 'Rated', 'Director', 'Writer', 'Actors', 'Plot', 'Language', 'Country']
    for col in string_columns:
        if col in df.columns:  # Check if column exists
            df[col] = df[col].str.strip()

    # Clean 'Runtime' column (convert to integer)
    if 'Runtime' in df.columns:
        df['Runtime'] = df['Runtime'].apply(lambda x: int(x.replace(' min', '')) if isinstance(x, str) else x)

    # Clean 'Metascore' column (convert to numeric)
    if 'Metascore' in df.columns:
        df['Metascore'] = pd.to_numeric(df['Metascore'], errors='coerce')

    # Clean 'Tomatometer Score' column (remove '%' character)
    if 'Tomatometer Score' in df.columns:
        df['Tomatometer Score'] = df['Tomatometer Score'].apply(
            lambda x: int(x.replace('%', '')) if isinstance(x, str) else x
        )

    # Clean 'imdbRating' column (remove everything from and after '/')
    if 'imdbRating' in df.columns:
        df['imdbRating'] = df['imdbRating'].apply(
            lambda x: x.split('/')[0].strip() if isinstance(x, str) and '/' in x else x
        )

    # Clean 'Metacritic' column (extract numeric score)
    if 'Metacritic' in df.columns:
        df['Metacritic'] = df['Metacritic'].apply(lambda x: int(x.split('/')[0]) if isinstance(x, str) else x)

    # Fill missing values for numerical columns
    for col in ['Runtime', 'Metascore', 'Tomatometer Score', 'Metacritic']:
        if col in df.columns:
            df[col] = df[col].fillna(df[col].median())

    # Save the cleaned data to a new CSV file
    df.to_csv(output_file, index=False)
    print(f"Cleaned data saved to {output_file}")

# Example usage
input_file = 'best_movies_27112024_055134.csv'  # Input CSV file path
output_file = 'cleaned_best_movies_27112024_055134.csv'  # Output cleaned CSV file path

# Main execution
if __name__ == "__main__":
    clean_csv_data(input_file, output_file)
```
